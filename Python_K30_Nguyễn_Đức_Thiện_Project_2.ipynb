# **Part II: Data Wrangling**

```python
#Using payment_report.csv & product.csv
#1. Top 3 product_ids with the highest volume
product_total_volume = payment_product.groupby("product_id")["volume"].sum().sort_values(ascending=False)
product_total_volume.iloc[:3] #Top 3 product_ids with the highest volume
```

```python
#2. Given that 1 product_id is only owed by 1 team, are there any abnormal products against this rule?
product[product.duplicated(subset="product_id")] #Không có sản phẩm nào xuất hiện 2 lần hay không có sản phẩm nào được sở hữu bởi 2 team trở lên
```

```python
#3.
#Find the team has had the lowest performance (lowest volume) since Q2.2023
data_since_Q2_2023 = payment_product[payment_product["report_month"] >= "2023-02"]
team_total_volume = data_since_Q2_2023.groupby("team_own")["volume"].sum().sort_values(ascending=True)
team_total_volume[:1] #APS is the team which has had the lowest performance (lowest volume) since Q2.2023
```

```python
#Find the category that contributes the least to that team
APS_product_payment_data = payment_product[payment_product["team_own"] == "APS"]
product_category_volume = APS_product_payment_data.groupby("category")["volume"].sum().sort_values(ascending=True)
product_category_volume[:1] #PXXXXXS is the category that contributes the least to that team
```

```python
#4. Find the contribution of source_ids of refund transactions (payment_group = ‘refund’), what is the source_id with the highest contribution?
refund_report = payment_report[payment_report["payment_group"] == "refund"]
total_refund_volume = refund_report["volume"].sum()
source_id_volume = refund_report.groupby("source_id")["volume"].sum().divide(total_refund_volume).sort_values(ascending=False)
source_id_volume[:1] #38 is the source_id with the highest contribution
```

```python
#Using transactions.csv
#1. Define type of transactions (‘transaction_type’) for each row
def transaction_type(transType, merchant_id):
  if transType == 2 and merchant_id == 1205:
    return "Bank Transfer Transaction"
  elif transType == 2 and merchant_id == 2260:
    return "Withdraw Money Transaction"
  elif transType == 2 and merchant_id == 2270:
    return "Top Up Money Transaction"
  elif transType == 2 and merchant_id != [1205, 2260, 2270]:
    return "Payment Transaction"
  elif transType == 8 and merchant_id == 2250:
    return "Transfer Money Transaction"
  elif transType == 8 and merchant_id != 2250:
    return "Split Bill Transaction"
  else:
    return "invalid transactions"
transactions["type_transaction"] = transactions.apply(lambda x: transaction_type(x["transType"], x["merchant_id"]), axis=1)
print(transactions.head())
```

```python
#2. Of each transaction type (excluding invalid transactions): find the number of transactions, volume, senders and receivers.
transaction_type_data = transactions[transactions["type_transaction"] != "invalid transactions"]
transaction_type_data.groupby("type_transaction").agg({"transaction_id": "count", "volume": "sum", "sender_id": "nunique", "receiver_id": "nunique"})
```
